histopgram grafiği

```python
import seaborn as sns 
diamonds=sns.load_dataset("diamonds")

df=diamonds.copy()
df.head
```

```python
#distplot sns içerisinden kaldırılmış. onun yerine displot /histplot ve kdeplot kullanabiliriz.
#displot: hem histogram hem KDE(kernel density estimate)
#histplot: sadece histogram 
#kdeplot:sadece KDE
sns.displot(df.price, kde=False); 
#bins ekleleyebiliriz. (aralık)

sns.displot(df.price, bins=100,kde=False); 
```

```python
?sns.displot
```

```python
sns.displot(df.price, kde=True); # kde true dersek  yoğunluğu ekliyor ve y ekseni 0-1 aralığnda yoğunluğa göre değişiyor
```

```python
sns.histplot(df.price) #sadece histogram
```

```python
?sns.histplot
```

```python
sns.histplot(df.price)
```

```python
sns.kdeplot(df.price, fill=True); #sadece yoğunluk. fill=True dedik ve altını da doldurduk
```

```python
# grafiğini geliştirelim 
g=sns.FacetGrid(diamonds, hue="cut")
g.map(sns.histplot,"price"); # hue=cut dersek alt kategorileri iç içe ekler ancak col=cut dersek ayrı ayrı xy eksenleri çizer

(sns.FacetGrid(df,
              hue="cut")).map(sns.histplot,
    "price")

```

```python
(sns.FacetGrid(df, 
               hue="cut",
              height=10 #eksen yüksekliği
              )).map(sns.histplot, "price" )
```

```python
(sns.FacetGrid(df, 
               hue="cut",
              height=5 ,#eksen yüksekliği
               xlim=(0,10000) #x eksen aralığı
              )).map(sns.histplot, "price" ).add_legend(); #add_legend diyerek alt grupların açıklamalarını ekledik
```

```python
#yukarıdakinin aynısını kde için yapalım 
(sns.FacetGrid(df, 
               hue="cut",
              height=5 ,#eksen yüksekliği
               xlim=(0,10000) #x eksen aralığı
              )).map(sns.kdeplot, "price" , shade=True).add_legend(); #add_legend diyerek alt grupların açıklamalarını ekledik

```
